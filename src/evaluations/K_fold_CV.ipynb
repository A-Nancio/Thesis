```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
global_state_df = pd.read_csv('../training_runs/global_state.csv')
feedzai_train_df = pd.read_csv('../training_runs/feedzaitrain.csv')
double_train_df = pd.read_csv('../training_runs/double_train.csv')
```

```python
df_list =  [global_state_df,  feedzai_train_df,  double_train_df]
for df in df_list:
    df['f1_score'] = df['f1_score'].apply(lambda x: eval(x)[0])
```

```python
global_state_df
```

```python
for df in df_list:
    df = df.groupby('epoch').mean()
```

```python
global_state_df = global_state_df.groupby('epoch').mean()
feedzai_train_df = feedzai_train_df.groupby('epoch').mean()
double_train_df = double_train_df.groupby('epoch').mean()

global_state_df.drop(['fold', 'f1_score'], inplace=True, axis=1)
feedzai_train_df.drop(['fold', 'f1_score'], inplace=True, axis=1)
double_train_df.drop(['fold', 'f1_score'], inplace=True, axis=1)

global_state_df['f1_score'] = 2 * (global_state_df['precision'] * global_state_df['recall']) / (global_state_df['precision'] + global_state_df['recall'])
feedzai_train_df['f1_score'] = 2 * (feedzai_train_df['precision'] * feedzai_train_df['recall']) / (feedzai_train_df['precision'] + feedzai_train_df['recall'])
double_train_df['f1_score'] = 2 * (double_train_df['precision'] * double_train_df['recall']) / (double_train_df['precision'] + double_train_df['recall'])
```
